```python
url = "https://www.worldometers.info/coronavirus/"
```

```python
import requests
```

```python
html = requests.get(url).text
```

```python
type(html)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
soup.find('tbody')
```

```python
tbody = soup.find('tbody')
```

```python
tr = tbody.find_all('tr')
```

```python
tr
```

```python
tr[8]
```

```python
for i in range(8, len(tr)):
    print(tr[i])
```

```python
complete_data = []
for i in range(8, len(tr)):
    country_data = []
    td= tr[i].find_all('td')  # one table row contains multiple table data
    for item in td:
        country_data.append(item.text) # extract data of each country
        
    complete_data.append(country_data) # append all data of each country on a list
```

```python
complete_data[0]
```

```python
clean_data = list(map(lambda x: x[1:10], complete_data))
clean_data
```

# Reading csv file in panda

```python
filename = 'covid-live_dec7.csv'
covid_data = pd.read_csv(filename)
covid_data.head()
```

```python
# Nan is missing value
```

# find no of missing value in column

```python
covid_data.isna() #True if missing else false
```

```python
missing  = covid_data.isna().sum()
```

```python
missing
```

```python
False +False
```

```python
True +True
```

```python
missing.plot.bar()
```

```python
type(missing)
```

# Accessinf index and values of pandas series

```python
missing.index
```

```python
missing.values
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.bar(x = missing.index, height = missing.values)
```

# increase fig size in matplotlib

```python
plt.figure(figsize =(20, 10))
plt.bar(x = missing.index, height = missing.values)
```

```python
plt.figure(figsize =(20, 10))
sns.barplot(x = missing.index, y = missing.values)
```

# missing data can also be shown on heatmap

```python
plt.figure(figsize= (12,5))
sns.heatmap(covid_data.isna().transpose())
```
